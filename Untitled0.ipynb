```python
import math
from prettytable import PrettyTable
import numpy as np
import matplotlib.pyplot as plt

#Задание функции f
def f(x):
    return x*x*math.cos(2*x)+1

#Задание и вывод итерационной таблицы
table = PrettyTable()
X = np.array([0,math.pi/4,math.pi/2,math.pi,4])
f_X = np.array([f(0),f(math.pi/4),f(math.pi/2),f(math.pi),f(4)])
table.field_names = ["x","0","pi/4","pi/2","pi","4"]
table.add_row(["f(x)",f_X[0],f_X[1],f_X[2],f_X[3],f_X[4]])
print("Рассмотрим заданную итерационную таблицу:")
print(table)

print("Задайте контрольную точку:")
x = float(input())

print("Задайте количество узлов:")
n = int(input())

#Построение равномерного распределения:
X_ravn = np.linspace(0,4,n)
Y_ravn = np.zeros(n)
for i in range(n):
  Y_ravn[i] = f(X_ravn[i])


#Выбор ближайших узлов из общего списка:
N = X.size
X_check = np.arange(N)*0;
X_new = np.zeros(n)
Y_new = np.zeros(n)
for i in range(n):
    minx = 100000
    minind = 0
    for j in range(N):
        if (abs(X[j]-x)<minx) and (X_check[j]==0):
            minx = abs(X[j]-x)
            minind = j
    X_check[minind] = 1
    X_new[i] = X[minind]
    Y_new[i] = f_X[minind]

#Вывод обрабатываемых узлов:
print("Выбраны узлы:")
table = PrettyTable()
Names = (np.arange(n)+1)
table.field_names = Names
table.add_row(X_new)
print(table)

#Интерполяция путем решения СЛАУ при Чебышевском распределении:
A = np.ones((n,n))
for i in range(n):
    for j in range(n):
        A[i][j]*= math.pow(X_new[i],j)
ans_SoLAE=np.linalg.solve(A,Y_new)

#Интерполяция путем решения СЛАУ при равномерном распределении:
AA = np.ones((n,n))
for i in range(n):
    for j in range(n):
        AA[i][j]*= math.pow(X_ravn[i],j)
ans_SoLAE_sec=np.linalg.solve(A,Y_ravn)

#Интерполяция при помощи полинома Лагранжа: нахождение множителей
def findML(x_this):
    Lmult = np.zeros(n)
    for i in range(n):
        l = 1.0
        for j in range(n):
            if j!=i:
                l = l * (x_this-X_new[j])
                l = l / (X_new[i]-X_new[j])
        Lmult[i] = l
    return Lmult

#Нахождение узлового члена
def wn(xx,nn):
    ans = 1
    for i in range(nn):
        ans = ans * (xx-X_new[i])
    return ans

#Нахождение узлового члена при равномерном распределении:
def wn_ravn(xx,nn):
    ans = 1
    for i in range(nn):
        ans = ans * (xx-X_ravn[i])
    return ans

#Интерпоялция при помощи метода Ньютона: нахождение разделенных разностей
RR_old = Y_new
RR_koef = np.zeros(n)
for i in range(n):
    RR_new = np.zeros(n-i-1)
    for j in range(0,n-i-1):
        RR_new[j] = (RR_old[j+1]-RR_old[j])/(X_new[j+i+1]-X_new[j])
    RR_koef[i] = RR_old[0]
    RR_old = RR_new

#Построение результатов:
X_g = np.linspace(0,4,100)
Y_g_true = np.zeros(100)
Y_g_SoLAE = np.zeros(100)
Y_g_errSoLAE = np.zeros(100)
Y_g_SoLAE_sec = np.zeros(100)
Y_g_errSoLAE_sec = np.zeros(100)
Y_g_L = np.zeros(100)
Y_g_errL = np.zeros(100)
Y_g_N = np.zeros(100)
Y_g_errN = np.zeros(100)
Wn = np.zeros(100)
Wn_ravn = np.zeros(100)
for i in range(100):
    Y_g_true[i]=f(X_g[i])
    #Вычисление узлового члена
    Wn[i] = wn(X_g[i],n)
    Wn_ravn[i] = wn_ravn(X_g[i],n)
    #Получение значений при решении СЛАУ
    ans = 0
    for j in range(n):
        ans+=math.pow(X_g[i],j)*ans_SoLAE[j]
    Y_g_SoLAE[i]=ans

    ans = 0
    for j in range(n):
        ans+=math.pow(X_g[i],j)*ans_SoLAE_sec[j]
    Y_g_SoLAE_sec[i]=ans

    #Получение значений при использовании полинома Лагранжа
    anss = 0
    LLmult = findML(X_g[i])
    for j in range(n):
        anss += LLmult[j]*Y_new[j]
    Y_g_L[i] = anss
    #Получение значений при использовании полинома Ньютона
    ans = 0
    for j in range(n):
        ans += RR_koef[j]*wn(X_g[i],j)
    Y_g_N[i] = ans
    
    #Получение значений погрешности
    if abs(Y_g_L[i]-Y_g_true[i])!=0:
        Y_g_errL[i] = -1*math.log10(abs(Y_g_L[i]-Y_g_true[i]))
    if abs(Y_g_SoLAE[i]-Y_g_true[i])!=0:
        Y_g_errSoLAE[i] = -1*math.log10(abs(Y_g_SoLAE[i]-Y_g_true[i]))
    if abs(Y_g_SoLAE_sec[i]-Y_g_true[i])!=0:
        Y_g_errSoLAE_sec[i] = -1*math.log10(abs(Y_g_SoLAE_sec[i]-Y_g_true[i]))
    if abs(Y_g_N[i]-Y_g_true[i])!=0:
        Y_g_errN[i] = -1*math.log10(abs(Y_g_N[i]-Y_g_true[i]))
print(X_ravn)
#Вывод графиков
fig = plt.subplots()
plt.plot(X_g,Y_g_errSoLAE,label = u'ошибка СЛАУ решения')
plt.plot(X_g,Y_g_errSoLAE_sec,label = u'ошибка СЛАУ решения при равномерном распределении')
plt.plot(X_g,Y_g_errL,label = u'ошибка Метода Лагранжа')
plt.plot(X_g,Y_g_errN,label = u'ошибка Метода Ньютона')
plt.legend()
plt.show()

fig = plt.subplots()
plt.plot(X_g,Wn,label = u'Узловой член при Чебышевском распределении')
plt.plot(X_g,Wn_ravn,label = u'Узловой член при равномерном распределении')
plt.legend()
plt.show()
```
